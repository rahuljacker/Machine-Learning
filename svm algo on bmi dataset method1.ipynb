```python
bmi["Gender"]=bmi["Gender"].map({"Male":1,"Female":0})
```

```python
bmi.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature=bmi.loc[:,"Gender":"Weight"]
target=bmi.loc[:,"Index"]
```

```python
target.head(3)
```

```python
featureTrain,featureTest,targetTrain,targetTest=train_test_split(feature,target,test_size=.30,random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
teacher=SVC(gamma=.001)
```

```python
learner=teacher.fit(featureTrain,targetTrain)
```

```python
learner.predict([[1,170,65]])
```

```python
from sklearn.metrics import accuracy_score
```

```python
Ya=targetTest
Yp=learner.predict(featureTest)
```

```python
acc=accuracy_score(Ya,Yp)*100
```

```python
feature=feature.values.reshape(len(feature),3)
target=target.values.reshape(len(target),1)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
featureTrain=feature[:370]
featureTest=feature[370:]
targetTrain=target[:370]
targetTest=target[370:]
```

```python
teacherGaussian=GaussianNB()
teacherBernouli=BernoulliNB()
teacherMultinomial=MultinomialNB()
teacherKnn=KNeighborsClassifier()
```

```python
learnerGaussian=teacherGaussian.fit(featureTrain,targetTrain)
learnerBernouli=teacherBernouli.fit(featureTrain,targetTrain)
learnerMultinomial=teacherMultinomial.fit(featureTrain,targetTrain)
learnerKnn=teacherKnn.fit(featureTrain,targetTrain)
```
